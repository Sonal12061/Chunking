```python
# Install required packages with specific versions
# Run this in your terminal or command prompt
# pip install --upgrade numpy pdf2image pypdf2

import numpy as np
from pdf2image import convert_from_path
from PyPDF2 import PdfReader, PdfWriter
import os

def extract_headings_from_pdf(pdf_path):
    """Extracts headings from a PDF by checking text properties."""
    reader = PdfReader(pdf_path)
    headings = []
    page_texts = []

    for i, page in enumerate(reader.pages):
        text = page.extract_text() or ""
        page_texts.append(text)

        for line in text.split("\n"):
            if len(line.strip()) > 0 and line.strip().isupper():  # Simple heading heuristic
                headings.append([line.strip(), i + 1])

    return headings, page_texts

def chunk_pdf_by_headings(pdf_path, headings):
    """Splits a PDF into sections based on detected headings and saves them as new PDFs."""
    reader = PdfReader(pdf_path)
    output_folder = "chunked_pdfs"
    os.makedirs(output_folder, exist_ok=True)

    num_pages = len(reader.pages)
    adjusted_headings = []

    # Adjust last heading if it is on the last page
    for i, (heading, start_page) in enumerate(headings):
        if i == len(headings) - 1 and start_page == num_pages:  # If last heading is on last page
            if adjusted_headings:
                adjusted_headings[-1][1] = start_page  # Merge with previous chunk
        else:
            adjusted_headings.append([heading, start_page])

    for i, (heading, start_page) in enumerate(adjusted_headings):
        end_page = adjusted_headings[i + 1][1] if i + 1 < len(adjusted_headings) else num_pages

        writer = PdfWriter()
        for j in range(start_page - 1, end_page):
            writer.add_page(reader.pages[j])

        output_path = os.path.join(output_folder, f"chunk_{i+1}_{heading[:20]}.pdf")
        with open(output_path, "wb") as output_pdf:
            writer.write(output_pdf)

    print(f"Chunked PDFs saved in '{output_folder}'")

# Example usage
pdf_path = FILE_PATH  # Replace with your PDF file
headings, _ = extract_headings_from_pdf(pdf_path)
chunk_pdf_by_headings(pdf_path, headings)
```
